```python
import matplotlib.pyplot as plt

#datos
x = [1,2,3,4,5]
y = [1,4,9,16,25]

#creamos el grafico
plt.plot(x,y)
plt.title("grafico de ejemplo")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

```python
categorias = ['A', 'B', 'C', 'D']
valores = [5, 7, 3, 8]

# crear el gr√°fico de barras
plt.bar(categorias, valores, color='green')
plt.title("Gr√°fico de Barras")
plt.xlabel("Categor√≠as")
plt.ylabel("Valores")
plt.show()
```
